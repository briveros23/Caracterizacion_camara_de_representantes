```python
import json
```

```python
with open('../input/partidos_a_estudiar.json', 'r') as archivo:
    diccionario_cargado = json.load(archivo)
```

```python
diccionario_cargado
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("somosnlp-hackathon-2022/paraphrase-spanish-distilroberta")
```

```python
leyes=[]
for key, value in diccionario_cargado.items():
    print(key,len(value))
    leyes += value
```

```python
embeddings = model.encode(leyes)
```

```python
import h5py
```

```python
with h5py.File('../input/embeddings.h5', 'w') as f:
    f.create_dataset('embeddings', data=embeddings)
```
